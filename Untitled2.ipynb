```python
import random as rand
import seaborn as sns
import numpy as np
```

```python
X = [rand.randint(0, 250) for _ in range(500)]
Y = [rand.randint(0, 250) for _ in range(500)]
print(X, Y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, Y, c='red')
plt.show()
```

```python
centers = []
for i in range(5):
  temp = []
  T = rand.randint(0, 499)
  temp.append(X[T])
  temp.append(Y[T])
  centers.append(temp)
centers
```

```python
D = [10000000000 for _ in range(0, 500)]
T = [0 for _ in range(0, 500)]
for _ in range(1000):
  for C_I, Temp in enumerate(centers):
      for P_I, point in enumerate(zip(X, Y)):
        distance = 0
        for i, j in zip(Temp, point):
            distance += (i-j) ** 2
        distance = distance ** 0.5
        if(D[P_I] > distance):
          T[P_I] = C_I
          D[P_I] = distance
  for i in range(5):
    centers[i][0] = np.mean([X[j] for j in range(0, 500) if T[j] == i])
    centers[i][1] = np.mean([Y[j] for j in range(0, 500) if T[j] == i])
```

```python
sns.scatterplot(X, Y, hue=T)
sns.scatterplot([i[0] for i in centers], [i[1] for i in centers], hue=[0,1,2,3,4], s=500)
```

```python
D = [100000 for _ in range(0, 500)]
T = [0 for _ in range(0, 500)]
for _ in range(1000):
  for C_I, Temp in enumerate(centers):
      for P_I, point in enumerate(zip(X, Y)):
        distance = 0
        for i, j in zip(Temp, point):
            distance += abs(j - i)
        if(D[P_I] > distance):
          T[P_I] = C_I
          D[P_I] = distance
  for i in range(5):
    centers[i][0] = np.mean([X[j] for j in range(0, 500) if T[j] == i])
    centers[i][1] = np.mean([Y[j] for j in range(0, 500) if T[j] == i])
```

```python
sns.scatterplot(X, Y, hue=T)
sns.scatterplot([i[0] for i in centers], [i[1] for i in centers], hue=[0,1,2,3,4], s=500)
```

```python
from sklearn.cluster import KMeans
import pandas as pd
```

```python
model = KMeans(n_clusters=5, max_iter=1000)
```

```python
df = pd.DataFrame()
df["X"] = X
df["Y"] = Y
df
```

```python
model.fit(df)
```

```python
answer = model.predict(df)
```

```python
C_lib = model.cluster_centers_
```

```python
sns.scatterplot(X, Y, hue=answer)
sns.scatterplot([i[0] for i in C_lib], [i[1] for i in C_lib], hue=[0,1,2,3,4], s=500)
```
